```python
11. Binary Operator Overloading using Friend
#include <iostream>
using namespace std;

class Add {
    int a;
public:
    Add(int x = 0) { a = x; }
    friend Add operator+(Add, Add);
    void show() { cout << "Value = " << a << endl; }
};

Add operator+(Add obj1, Add obj2) {
    return Add(obj1.a + obj2.a);
}

int main() {
    Add a1(5), a2(10), a3;
    a3 = a1 + a2;
    a3.show();
    return 0;
}
```

```python
4. Scope Resolution Operator
#include <iostream>
using namespace std;

int x = 10;  // Global variable

int main() {
    int x = 20;  // Local variable
    cout << "Local x = " << x << endl;
    cout << "Global x = " << ::x; // Using scope resolution
    return 0;
}
```

```python
10. Deep Copy Constructor
#include <iostream>
#include <cstring>
using namespace std;

class Student {
    char* name;
public:
    Student(const char* n) {
        name = new char[strlen(n) + 1];
        strcpy(name, n);
    }
    Student(const Student& s) {  // Deep copy
        name = new char[strlen(s.name) + 1];
        strcpy(name, s.name);
    }
    void show() { cout << "Name: " << name << endl; }
    ~Student() { delete[] name; }
};

int main() {
    Student s1("John");
    Student s2 = s1;
    s1.show();
    s2.show();
    return 0;
}
```

```python
8. Object as Parameter and Return Type
#include <iostream>
using namespace std;

class Number {
    int n;
public:
    void set(int x) { n = x; }
    void show() { cout << "Value = " << n << endl; }
    Number add(Number obj) {
        Number temp;
        temp.n = n + obj.n;
        return temp;
    }
};

int main() {
    Number n1, n2, n3;
    n1.set(5);
    n2.set(10);
    n3 = n1.add(n2);
    n3.show();
    return 0;
}
```

```python
16. Function Template (Max of Two Values)
#include <iostream>
using namespace std;

template <class T>
T maxValue(T a, T b) {
    return (a > b) ? a : b;
}

int main() {
    cout << "Max: " << maxValue(10, 20) << endl;
    cout << "Max: " << maxValue(3.5, 2.8);
    return 0;
}
```

```python
12. Single Inheritance
#include <iostream>
using namespace std;

class A {
public:
    void displayA() { cout << "Base Class\n"; }
};

class B : public A {
public:
    void displayB() { cout << "Derived Class\n"; }
};

int main() {
    B obj;
    obj.displayA();
    obj.displayB();
    return 0;
}
```

```python
2. Factorial using Recursion
#include <iostream>
using namespace std;

int factorial(int n) {
    if (n <= 1)
        return 1;
    return n * factorial(n - 1);
}

int main() {
    int n = 5;
    cout << "Factorial of " << n << " = " << factorial(n);
    return 0;
}
```

```python
18. Re-throwing Exceptions
#include <iostream>
using namespace std;

void test() {
    try {
        throw "Error occurred!";
    } catch (const char* msg) {
        cout << "Caught inside function. Rethrowing...\n";
        throw; // Re-throw
    }
}

int main() {
    try {
        test();
    } catch (const char* msg) {
        cout << "Caught in main: " << msg;
    }
    return 0;
}
```

```python
3. Factorial using Loops
#include <iostream>
using namespace std;

int main() {
    int n = 5, fact = 1;
    for (int i = 1; i <= n; i++)
        fact *= i;
    cout << "Factorial of " << n << " = " << fact;
    return 0;
}
```

```python
15. Bubble Sort using Function Template
#include <iostream>
using namespace std;

template <class T>
void bubbleSort(T arr[], int n) {
    for (int i = 0; i < n - 1; i++)
        for (int j = 0; j < n - i - 1; j++)
            if (arr[j] > arr[j + 1])
                swap(arr[j], arr[j + 1]);
}

int main() {
    int arr[5] = {5, 3, 4, 1, 2};
    bubbleSort(arr, 5);
    cout << "Sorted: ";
    for (int i = 0; i < 5; i++) cout << arr[i] << " ";
    return 0;
}
```

```python
6. Friend Function
#include <iostream>
using namespace std;

class Sample {
    int num;
public:
    Sample(int n) : num(n) {}
    friend void show(Sample s);
};

void show(Sample s) {
    cout << "Number = " << s.num;
}

int main() {
    Sample obj(10);
    show(obj);
    return 0;
}
```

```python
14. Pure Virtual Function / Abstract Class
#include <iostream>
using namespace std;

class Shape {
public:
    virtual void area() = 0;  // Pure virtual
};

class Square : public Shape {
public:
    void area() { cout << "Area of Square = side * side\n"; }
};

class Triangle : public Shape {
public:
    void area() { cout << "Area of Triangle = 0.5 * base * height\n"; }
};

int main() {
    Square s;
    Triangle t;
    s.area();
    t.area();
    return 0;
}
```

```python
7. Array of Objects
#include <iostream>
using namespace std;

class Student {
public:
    int id;
    string name;
    void getData() {
        cout << "Enter ID and Name: ";
        cin >> id >> name;
    }
    void showData() {
        cout << id << " " << name << endl;
    }
};

int main() {
    Student s[3];
    for (int i = 0; i < 3; i++)
        s[i].getData();

    cout << "\nStudent Details:\n";
    for (int i = 0; i < 3; i++)
        s[i].showData();
    return 0;
}
```

```python
13. Virtual Base Class
#include <iostream>
using namespace std;

class A {
public:
    void showA() { cout << "Class A\n"; }
};

class B : virtual public A {};
class C : virtual public A {};
class D : public B, public C {};

int main() {
    D obj;
    obj.showA();
    return 0;
}
```
